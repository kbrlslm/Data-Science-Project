```python
# Create Data frame ans see the data
data = pd.read_csv('NYPD_Arrests_Data.csv')
data.head()
```

```python
# find NA value of each column
data.isnull().sum()
```

```python
# delete unnecessary data for the project
data.drop(['ARREST_KEY','PD_CD','PD_DESC','KY_CD','LAW_CODE',
            'JURISDICTION_CODE','X_COORD_CD','Y_COORD_CD','Latitude','Longitude', 'Lon_Lat'],axis=1,inplace=True)
data
```

```python
# copy the dataset
data_clean = data.copy()
```

```python
    # convert to datetime object
    data_clean['ARREST_DATE'] = pd.to_datetime(data_clean['ARREST_DATE'])
```

```python
# drop any rows with nulls
data_clean=data_clean.dropna()
```

```python
# sort data by date
data_clean.sort_values(by=['ARREST_DATE'], inplace=True)
```

```python
# Rename the attribute name
data_clean['OFFENCE_LEVEL'] = data_clean['LAW_CAT_CD'].map({ 'F': 'Felony',
                                                                    'M': 'Misdemeanor',
                                                                    'V': 'Violation',
                                                                    'I': 'Infractions' })
```

```python
# Rename the Borough
data_clean['ARREST_BORO'] = data_clean['ARREST_BORO'].map({ 'K': 'Brooklyn',
                                                                'B': 'Bronx',
                                                                'M': 'Manhattan',
                                                                'Q': 'Queens',
                                                                'S': 'Staten Island' })
```

```python
data_clean.head()
```

```python
# Drop 'LAW_CAT_CD'
data_clean.drop(['LAW_CAT_CD'],axis=1,inplace=True)
```

```python
#Export the dataset
data_clean.to_csv('data_clean.csv', index = False)
```
