```python
data['ARREST_DATE']=pd.to_datetime(data['ARREST_DATE'])
```

```python
arrest_year=data['ARREST_DATE'].dt.year
len(arrest_year)
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
arrest_year.value_counts(ascending=True)
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x=arrest_year, palette='dark:salmon_r')
_ = plt.title('The Number of Arrests by Year')
_ = plt.xlabel('Arrest By Year')
_ = plt.ylabel('Arrest Count')
```

```python
plt.figure(figsize=(30,20))
sns.countplot(x=arrest_year,hue=data['ARREST_BORO'],palette='icefire')
_ = plt.title('The Number of Arrests Per Year by Borough',fontsize=25)
_ = plt.xlabel('Arrest By Year',fontsize=25)
_ = plt.ylabel('Arrest Count',fontsize=25)
_ = plt.legend(loc="upper right",fontsize=25)
_ = plt.xticks(fontsize=20)
_ = plt.yticks(fontsize=20,rotation=55)
```

```python
plt.figure(figsize=(30,20))
sns.countplot(x=data['ARREST_BORO'],hue=data['LAW_CAT_CD'],palette='viridis')
_ = plt.title('The Number of Arrests by Borough Based on Offense Level',fontsize=25)
_ = plt.xlabel('Name of Borough',fontsize=25)
_ = plt.ylabel('Arrest by Count',fontsize=25)
_ = plt.legend(loc="upper right",fontsize=25)
_ = plt.xticks(fontsize=20)
_ = plt.yticks(fontsize=20,rotation=55)
```

```python
plt.figure(figsize=(30,20))
sns.countplot(x=arrest_year,hue=data['AGE_GROUP'],hue_order=['<18', '18-24', '25-44', '44-65', '65+'], palette='icefire')
_ = plt.title('The Number of Arrests Per Year by Age Group',fontsize=25)
_ = plt.xlabel('Arrests per Year',fontsize=25)
_ = plt.ylabel('Arrest Count',fontsize=25)
_ = plt.legend(loc="upper right",fontsize=25)
_ = plt.xticks(fontsize=20)
_ = plt.yticks(fontsize=20,rotation=55)
```
